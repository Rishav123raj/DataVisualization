```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(color_codes=True)
%matplotlib inline
```

```python
auto = pd.read_csv('Automobile.csv')
```

```python
auto.head()
```

```python
sns.distplot(auto['normalized_losses'])
plt.show()
```

```python
sns.distplot(auto['normalized_losses'], kde=False, rug=True)
plt.show()
```

```python
sns.jointplot(data=auto, x='engine_size', y='horsepower', kind='kde')
plt.show()
```

```python
sns.pairplot(auto[['normalized_losses','engine_size','wheel_base']])
```

```python
sns.swarmplot(data=auto, x='fuel_type', y='horsepower')
```

```python
sns.boxplot(data=auto, x='number_of_doors', y='horsepower', hue=auto['fuel_type'])
```

```python
sns.countplot(data=auto, x='body_style')
```

```python
sns.pointplot(data=auto, x='fuel_system', y='horsepower')
```

```python
sns.lmplot(data=auto, y='peak_rpm', x='horsepower', hue='fuel_type')
```
